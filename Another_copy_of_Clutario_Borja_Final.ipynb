```python
print(df.shape)
# Returns (rows, columns) - In this case (5, 12) based on the sample provided
```

```python
print(df.head())  # First five rows
print(df.tail())  # Last five rows
```

```python
print(df.info())
```

```python
print(df.isnull().sum())
```

```python
df['director'].fillna('Unknown', inplace=True)
df['cast'].fillna('Unknown', inplace=True)
df['country'].fillna('Unknown', inplace=True)
df['date_added'].fillna(df['date_added'].mode()[0], inplace=True)
```

```python
print(df.dtypes)
```

```python
df['date_added'] = pd.to_datetime(df['date_added'])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.countplot(x='type', data=df)
plt.title('Distribution of Movies vs TV Shows')
plt.show()
```

```python
print(df['rating'].value_counts().head(5))
```

```python
print(df['release_year'].value_counts().head(1))
```

```python
df['duration_numeric'] = df['duration'].str.extract('(\d+)').astype(int)
df['duration_type'] = df['duration'].str.extract('([A-Za-z]+)')
```

```python
avg_movie_length = df[df['type'] == 'Movie']['duration_numeric'].mean()
print(f'Average Movie Length: {avg_movie_length:.2f} minutes')
```

```python
avg_seasons = df[df['type'] == 'TV Show']['duration_numeric'].mean()
print(f'Average Seasons for TV Shows: {avg_seasons:.2f} seasons')
```

```python
df['first_genre'] = df['listed_in'].apply(lambda x: x.split(',')[0])
genre_release_year = df.groupby('first_genre')['release_year'].mean().sort_values(ascending=False)
print(genre_release_year)
```

```python
df['year_added'] = df['date_added'].dt.year
release_vs_added = df.groupby('release_year')['year_added'].mean()
plt.figure(figsize=(12, 6))
plt.plot(release_vs_added.index, release_vs_added.values)
plt.xlabel('Release Year')
plt.ylabel('Average Year Added to Netflix')
plt.title('Temporal Trend: Release Year vs. Year Added')
plt.show()
```

```python
cross_tab = pd.crosstab(df['type'], df['rating'])
print(cross_tab)
```

```python
r_rated_after_2020 = df[(df['rating'] == 'R') & (df['year_added'] > 2020)]
print(f"Number of R-Rated Titles Added After 2020: {len(r_rated_after_2020)}")
print(r_rated_after_2020['title'])
```

```python
avg_release_year_by_country = df.groupby('country')['release_year'].mean().sort_values(ascending=False)
print(avg_release_year_by_country)
```

```python
type_by_year = df.groupby(['year_added', 'type']).size().unstack(fill_value=0)
type_by_year['total'] = type_by_year['Movie'] + type_by_year['TV Show']
type_by_year['movie_proportion'] = type_by_year['Movie'] / type_by_year['total']
print(type_by_year[['movie_proportion']])
```
